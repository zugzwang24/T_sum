```python
import pandas as pd
import numpy as np

# 1. CSV 파일 읽기 
file_path = '유동인구.csv' 
try:
    df = pd.read_csv(file_path, encoding='cp949')
    print("데이터 로드 성공 (cp949)")
except:
    df = pd.read_csv(file_path, encoding='utf-8')
    print("데이터 로드 성공 (utf-8)")
```

```python
# 2. 데이터 상위 5행 확인 (객체의 머리 보기) head() 함수는 5행을 기본으로 보여줍니다.
print(df.head())

# 3. 데이터의 차원 확인 (행, 열 개수) shape 속성은 (행, 열) 형태로 반환합니다.
print(f"전체 행 개수: {df.shape[0]}, 전체 열(항목) 개수: {df.shape[1]}")

# 4. 데이터 타입 및 결측치 확인 (데이터 정합성 체크) 결측치란 데이터가 없는 부분을 의미합니다.
# 여기서 'Dtype'이 object면 문자열, float/int면 숫자입니다.
# 숫자로 계산해야 할 매출액이 object라면 나중에 형변환이 필요함을 배울 수 있습니다.
print(df.info())    #info() 함수는 각 열의 데이터 타입과 결측치 개수를 보여줍니다.
```

```python
# '행정동_코드_명'이 '강남역'이나 '역삼1동'인 데이터만 필터링 (본인 동네로 수정 가능)
my_dong = df[df['행정동_코드_명'] == '역삼1동']
print(my_dong[['기준_년분기_코드', '행정동_코드_명', '총_유동인구_수']])
```

```python
# 총_유동인구_수 기준 내림차순 정렬
# 주의: 데이터 타입이 숫자(int/float)인지 확인이 필요할 수 있습니다.
top_10 = df.sort_values(by='총_유동인구_수', ascending=False).head(10)
print(top_10[['행정동_코드_명', '총_유동인구_수']])
```
